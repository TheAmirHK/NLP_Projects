```python
import openai
import edge_tts
import asyncio
import nest_asyncio
import IPython.display as ipd

nest_asyncio.apply()

openai.api_key = "OpenAI key"

def get_book_summary(title, author):
    prompt = f"Provide a long summary of the book '{title}' by {author}. if the author is not exactly the same just correct it and don't apologize. just give the summary"

    client = openai.OpenAI(api_key=openai.api_key)

    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "You are a helpful assistant."},
                  {"role": "user", "content": prompt}]
    )

    summary = response.choices[0].message.content
    return summary

async def speak_summary(summary, voice="en-GB-RyanNeural"):
    output_file = "summary.mp3"
    communicate = edge_tts.Communicate(summary, voice=voice)
    await communicate.save(output_file)
    return output_file

def play_audio(audio_file):
    ipd.display(ipd.Audio(audio_file))

if __name__ == "__main__":
    book_title = input("Enter the book title: ")
    book_author = input("Enter the author's name: ")

    summary = get_book_summary(book_title, book_author)
    print("\nBook Summary:\n", summary)

    loop = asyncio.get_event_loop()
    audio_file = loop.run_until_complete(speak_summary(summary))

    play_audio(audio_file)
```
